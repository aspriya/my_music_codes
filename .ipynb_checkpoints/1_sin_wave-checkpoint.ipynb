```python
#Equation for a simple cos wave is as follows.
# x(n) = A * cos(2 * pi * f * n * T + phi)

import matplotlib.pyplot as plt
import numpy as np

A = .8
f0 = 1000
phi = np.pi/2
fs = 44100
t = np.arange(-.002, .002, 1.0/fs)   #very important function in numpy. arange can form a array of values
x = A * np.cos(2 * np.pi * f0 * t + phi)

plt.plot(t, x)
plt.axis([-.002,.002, -.8, .8])
plt.xlabel('time')
plt.ylabel('amplitude')
plt.show()
```
