```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
import scipy.io.wavfile as wavfile

M = 501
(fs, signal_samples) = wavfile.read('orchestra.wav')
x1 = signal_samples[len(signal_samples)/2 : len(signal_samples)/2 + M] * np.hamming(M)
plt.plot(x1)
plt.show()
```

```python
N = 1024  #a power of two number after to 501
fftBuffer = np.zeros(N)
fftBuffer[0:251] = x1[250:501]
fftBuffer[N-249:] = x1[0:249]
plt.plot(fftBuffer)
plt.show()
```

```python
X = fft(fftBuffer)
mX = abs(X)      
#The absolute value of a complex number (also called as the modulus) is defined as 
#the distance between the origin and the point in the complex plane.

pX = np.angle(X)
plt.plot(mX[0:512])  #we need to plot only half of the magnitude spectrum as other half is just a reflrection.
plt.show()
```

```python

mX = 20*np.log10(abs(X))
plt.plot(mX[0:512])
plt.show()
```
