```python
# Equation for DFT operation is as follows
# X(k) = sigma_n_0_to_N-1(x(n). exponential(-j2.pi.k.n/N)) ; k = 0,1,2,..., N-1

# In DFT we find coificient values of all X(k). So we are iterating through k.

import matplotlib.pyplot as plt
import numpy as np
import random

#Creating a random input signal
fs = 44100
a = float(random.randint(1, 100))
f0 = float(random.randint(500, 1000))
p = float(random.randint(1, 360))

t = np.arange(0, .004, 1.0/fs)
fnList = a * np.sin(2 * np.pi * f0 * t + p / 360 )

print("Number of samples:"+str(len(t))+" a:"+str(a)+" f0:"+str(f0)+" p:"+str(p))
# y(t) = A.sin(2.pi.f.t + phase)

plt.plot(t, fnList)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

```python
#Fuction to calculate the DFT once a set of samples of a input signal is given.
def DFT(fnList):
    N = len(fnList)
    FkList = []
    for k in range(N):
        Fk = 0.0
        for n in range(N):
            Fk += fnList[n] * np.exp(- 1j * 2 * np.pi * k * n / N)
        FkList.append(Fk)
    return FkList


freq_list = DFT(fnList)
plt.plot(t, np.real(freq_list))
plt.show()
```
