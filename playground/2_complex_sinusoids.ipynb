```python
#Equation for a complex cos wave is as follows.
# sk(n) = exponential(j2.pi.k.n/N) = cos(2.pi.k.n/N) + jsin(2.pi.k.n/N)

import matplotlib.pyplot as plt
import numpy as np

N = 500
k = 3
n = np.arange(-N/2, N/2)

s = np.exp(1j * 2 * np.pi * k * n / N)

plt.plot(n, np.real(s)) # as from a complex sinusoid we get two parts as real and imaginary, we cant plot
                        # both of them and n values. Because then we have 3 dimensions. so cant plot using a
                        # 2D diagram. That is why here we only plot n and real part of s

plt.axis([-N/2, N/2-1, -1, 1])
plt.xlabel('n')
plt.ylabel('amplitude')
plt.show()
```
